All that remains now is to gather all the scanned files into one merged dataframe.

```python
#create an empty dataframe for all Cyclistic rides
rides = pd.DataFrame()

#this will aggregate row count of all files to compare with the final row count of the rides dataframe
row_count = 0
#this is the same code for extracting each file from '2023_Cyclistic_Trips'
for month in months:
  df = pd.read_csv(pathstart+month+pathend)
  #add row count each table to row_count
  row_count += len(df)
  #an additional line of code will be added to merge each file into the rides dataframe
  rides = pd.concat([rides,df], axis=0, ignore_index=True)

#check for total row count of all tables match total row count in the merged rides dataframe
print(row_count)
len(rides)
```

The `row_count` total and the `rides` dataframe length are both the same meaning all row values were carefully appended into the aggregate dataframe. The `concat()` function served most effective for merging the tables because all of them contained the exact same column names. Using `axis=0` allowed the tables to stack the first dataframe above the next dataframe without losing or adding any new fields or values. This is comparable to the `Union All` clause in SQL as opposed to `Join` because it merges two or more tables with matching columns vertically. The `Join` clause will add columns horizontally and fill in empty row values with *null* placeholders depending on the join type. The new table contains nearly 6 MILLION rows of data. ###Supportive Data Transformations

A crucial next step in this stage will be to extract the valuable trip duration of each ride stored by the start and end at columns. To do this the analyst must start off by changing the format of the datetime columns from string objects to datetime objects. This will allow for smooth calculations in time differences between the columns.

```python
#quick check of the data type stored in 'started_at' and 'ended_at' columns
print(type(clean_rides.started_at[0]))
print(type(clean_rides.ended_at[0]))

#change format to datetime 'Timestamps' using .to_datetime() for each column
clean_rides['started_at'] = pd.to_datetime(clean_rides['started_at'])
clean_rides['ended_at'] = pd.to_datetime(clean_rides['ended_at'])

#check format again
print(type(clean_rides.started_at[0]))
print(type(clean_rides.ended_at[0]))
```

Now the columns are of 'Timestamp' format allowing it to be treated as time instead of a line of text. This will allow performing certain arithmetic operations on the datetime objects without returning an error. For this analysis the analyst will find the difference between the end time and start time for the total trip duration. For ease of view this will be inserted right next to the original time columns.

```python
#use the insert() function to add column 'duration' as the fifth column (index 4)
#the values will be the calculated difference of the 'started_at' column from the 'ended_at' column
#additionally the total duration time will be converted into seconds using dt.total_seconds()
clean_rides.insert(4, 'duration', (clean_rides['ended_at'] - clean_rides['started_at']).dt.total_seconds())

#make sure to not run this code block more than once as will try to make a new insert each time
clean_rides.head(3)
```

Excellent! The table is now showing the duration between the start and end times at the fifth column. This number represents the total seconds of the ride. It can be more meaningful to the audience if the duration is in minutes. This can be changed with a simple calculation.

```python
#divide the values of the duration columns be 60 and round two decimal places for duration in minutes
clean_rides['duration'] = round(clean_rides['duration']/60, 2)

#rename the 'duration' column to 'total_time'
clean_rides = clean_rides.rename(columns={"duration":"total_time"})
#RUN ONLY ONCE! or the calculation will keep dividing by 60 from the first line
clean_rides.head(5)
```

The dataset is almost ready for analysis. There is one more column that could be of use to calculate into the table. This is the day of the week that each ride had started on. This step is already half done by formating the datetime columns into 'Timestamp' objects. The new column will be deriving the day of the week from the `'started_at'` column and placed sixth in the column order.

```python
#the insert() function will insert the 'day_of_week' column as the sixth row
clean_rides.insert(5, 'day_of_week', clean_rides['started_at'].dt.day_name())

#RUN ONLY ONCE!
clean_rides.head(3)
```
